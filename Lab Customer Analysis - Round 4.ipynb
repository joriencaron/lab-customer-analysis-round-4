```python
#Round 4 starts here

#transform the data to numerical and categorical
numerical = df_marketing.select_dtypes(include= np.number)
numerical
```

```python
numerical = numerical.drop(['Unnamed: 0'], axis=1)
```

```python
categoricals = df_marketing.select_dtypes(include= np.object)
categoricals
```

```python
#check the normality of the numerical variables
for column in numerical.columns:
    sns.displot(x=column, data=numerical, bins=30)
    plt.show()

#in my perspective none of them look normally distributed
```

```python
#check the normality of the numerical variables
for column in numerical.columns:
    plt.hist(x=column, data=numerical, bins=30)
    plt.show()
```

```python
#check the multicollinearity
correlation_matrix = numerical.corr()

correlation_matrix
```

```python
#check the multicollinearity
plt.figure(figsize=(12, 9))
sns.heatmap(correlation_matrix, annot=True)
plt.show()

#there is no pair with high correlation above .90. The heatmap does show us that 'Total claim amount' has interesting 
#correlation with 'Customer lifetime value' 'Income' and 'Monthly premium auto'. This will be interesting for further analysis.
#the other columsn might need to be dropped, but I think I can test that in the upcoming labs
```
